```python
import random

def com(combinacoes):
    print("Você gostaria de escolher uma dessas combinações automaticamente? (sim/não)")
    escolha = input().lower()
    if escolha == "sim" and combinacoes:
        senha_aleatoria = random.choice(combinacoes)
        print(f"Sua senha escolhida automaticamente é: {senha_aleatoria}")
    elif not combinacoes:
        print("Nenhuma combinação disponível.")
    else:
        print("Você optou por não selecionar uma combinação automaticamente.")

def rep(quantidade):
    senha = []
    for i in range(quantidade):
        numero = int(input(f"Digite o número {i + 1}: "))
        senha.append(numero)
    return senha

senha = []
combinacoes = []

quantidade_de_digitos_da_senha = int(input("Quantos caracteres tem sua senha (EX: 2 = '12'): "))

if quantidade_de_digitos_da_senha == 2:
    senha = rep(quantidade_de_digitos_da_senha)

    print(f"Temos:\n2 opções para o primeiro dígito  ({', '.join(map(str, senha))}).\n"
          f"2 opções para o segundo dígito  ({', '.join(map(str, senha))}).\n"
          f"Assim, o número total de combinações possíveis é 2×2 = 4.")

    for i in senha:
        for o in senha:
            combinacao = f"{i}{o}"
            combinacoes.append(combinacao)
            print(combinacao)

    com(combinacoes)

elif quantidade_de_digitos_da_senha == 3:
    repetidor = input("Seus números vão poder se repetir? (sim/não): ").lower()
    senha = rep(quantidade_de_digitos_da_senha)

    print(f"Temos:\n3 opções para o primeiro dígito ({', '.join(map(str, senha))}).\n"
          f"3 opções para o segundo dígito ({', '.join(map(str, senha))}).\n"
          f"3 opções para o terceiro dígito ({', '.join(map(str, senha))}).\n"
          f"Assim, o número total de combinações possíveis é 3×3×3 = 27.")

    if repetidor == "sim":
        for i in senha:
            for o in senha:
                for p in senha:
                    combinacao = f"{i}{o}{p}"
                    combinacoes.append(combinacao)
                    print(combinacao)
    else:
        for i in senha:
            for o in senha:
                for p in senha:
                    if len(set([i, o, p])) == 3:
                        combinacao = f"{i}{o}{p}"
                        combinacoes.append(combinacao)
                        print(combinacao)

    com(combinacoes)

elif quantidade_de_digitos_da_senha == 4:
    separados = input("Seus números vão poder se repetir? (sim/não): ").lower()
    senha = rep(quantidade_de_digitos_da_senha)

    print(f"Temos:\n4 opções para o primeiro dígito ({', '.join(map(str, senha))}).\n"
          f"4 opções para o segundo dígito ({', '.join(map(str, senha))}).\n"
          f"4 opções para o terceiro dígito ({', '.join(map(str, senha))}).\n"
          f"4 opções para o quarto dígito ({', '.join(map(str, senha))}).")

    if separados == "sim":
        for i in senha:
            for o in senha:
                for p in senha:
                    for h in senha:
                        combinacao = f"{i}{o}{p}{h}"
                        combinacoes.append(combinacao)
                        print(combinacao)
    else:
        for i in senha:
            for o in senha:
                for p in senha:
                    for h in senha:
                        if len(set([i, o, p, h])) == 4:
                            combinacao = f"{i}{o}{p}{h}"
                            combinacoes.append(combinacao)
                            print(combinacao)

    com(combinacoes)

else:
    print("Número de dígitos inválido! Escolha entre 2, 3 ou 4.")
```
